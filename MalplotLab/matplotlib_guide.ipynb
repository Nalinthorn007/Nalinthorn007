#### Step 3: graph (Bar Chart)

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
# difine x and y 
# x = np.array([2018,2019,2020,2021,2022,2023])
# y = np.array([1050,2452.20,3456.30,4653.40,4052.60,2233.45])
df = pd.read_csv('../pandasLab/dataset/gdp.csv')
plt.bar(df['year'],df['gdp'] , color=['green','blue','black','yellow','red','lightblue'], edgecolor='black')
plt.title('Laos GDP')
plt.xlabel('Year')
plt.ylabel('GDP (Million kip)')
plt.show
```
